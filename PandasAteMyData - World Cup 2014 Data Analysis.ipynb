## Visualization

Show a Pie Chart all the Brazilian's age group

```python
#playersAge = '(1925-06-27)27 June 1925 (aged z)'
#playersAge = '(1925-06-27)27 June 1925 (aged 18)'
#playersAge = '(1925-06-27)27 June 1925 (aged 20)'
#playersAge = '(1925-06-27)27 June 1925 (aged 31)'

playersAge = ''

def representsInt(s):
    try: 
        int(s)
        return True
    except ValueError:
        return False

def determineAge(age_group):
    if age_group != '':
        if isinstance(age_group, str):
            arr = age_group.split('(aged ')
            arr2 = arr[1].split(')')
            if representsInt(arr2[0]) :
                realAge  = int(arr2[0])
                if realAge <= 19:
                    return "Young got potential"
                if realAge <= 30 and realAge > 19:
                    return "Career start here !"
                else:
                    return "Above 30 better score more before retirement !"

    
determineAge(playersAge)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use("ggplot")

brazildf["DOB/Age"].apply(determineAge).value_counts().plot(kind = "pie", legend = True)
```
